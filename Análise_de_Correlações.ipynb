## Execução de funções necessárias

```python
## Estratégia para tratar múltiplas respostas
def multiplas(respostas):
  quebradas = ['']
  for i in respostas:
      v = i.split(';')
      if(len(v)==1):
          quebradas.append(v[0])
      else:        
          for j in v:
              quebradas.append(j)
  return quebradas
```
